```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = ["Burj Khalifa", "Shanghai Tower", "Abraj Al-Bait Clock Tower",
     "Ping An Finance Center", "Lotte World Tower", "One World Trade Center",
     "Guangzhou CTF Finance Center", "Tianjin CTF Finance Center",
     "China Zun", "Taipei 101"]
y = [828, 632, 601, 599, 554.5, 541.3, 530, 530, 528, 508]
plt.style.use("ggplot")
plt.bar(x, y)
plt.xticks(rotation=70)
plt.title("Height by Building")
plt.xlabel("Building")
plt.ylabel("Height (m)")
plt.savefig("height.png", bbox_inches='tight')
```
